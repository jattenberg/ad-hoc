```python
import pandas as pd
from matplotlib import pyplot as plt
import os
from nameparser import HumanName
import datetime as dt
```

```python
dir = "/Users/jattenberg/Downloads/noahs-csv/"
files = os.listdir(dir)
names = [n.split(".")[0] for n in files]
dataframe_arr = [pd.read_csv(dir+f) for f in files]

dfs = dict(zip(names, dataframe_arr))
```

```python
products = dfs["noahs-products"]
order_items = dfs["noahs-orders_items"]
orders = dfs["noahs-orders"]
customers = dfs["noahs-customers"]
```

```python
queens_village_customers = customers[customers.citystatezip.str.contains("queens village", case=False)]
queens_village_customers
```

```python
cat_items = products[products.desc.str.contains("cat", case=False, regex=True)]
cat_order_items = order_items.merge(cat_items, on="sku").drop_duplicates("orderid")
cat_orders = cat_order_items.merge(orders, on="orderid")
cat_orders
```
