```python
import pandas as pd
from matplotlib import pyplot as plt
import os
from nameparser import HumanName
```

```python
dir = "/Users/jattenberg/Downloads/noahs-csv/"
os.listdir(dir)
```

```python
customers = pd.read_csv(dir+"noahs-customers.csv")
customers
```

```python
customers["last_name"] = customers.name.apply(lambda x: HumanName(x).last.lower())
customers["digits"] = customers.phone.apply(lambda x: "".join(x.split("-")[-3:]))
customers
```

```python
keypad = {
    2: ["A", "B", "C"],
    3: ["D", "E", "F"],
    4: ["G", "H", "I"],
    5: ["J", "K", "L"],
    6: ["M", "N", "O"],
    7: ["P", "Q", "R", "S"],
    8: ["T", "U", "V"],
    9: ["W", "X", "Y", "Z"],
}
letter_mapping = dict((n.lower(), str(k)) for k,ns in keypad.items() for n in ns)

def number_to_strings(s):
    "the number is actually a string"
    return "_".join(",".join(keypad.get(int(x), ["$"])) for x in s)

def string_to_number(s):
    return "".join([letter_mapping[x] for x in s])

customers["name_as_num"] = customers.last_name.apply(string_to_number)
customers
```

## What are the cases where the number-mapped last name matches the last 4 of the phone number? 

```python
def numbers_match(r):
    return r.digits == r.name_as_num
customers[customers.apply(numbers_match, axis=1)]
```
