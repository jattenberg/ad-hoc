```python
import pandas as pd
from matplotlib import pyplot as plt
import os
from nameparser import HumanName
import datetime as dt
```

```python
dir = "/Users/jattenberg/Downloads/noahs-csv/"
files = os.listdir(dir)
names = [n.split(".")[0] for n in files]
dataframe_arr = [pd.read_csv(dir+f) for f in files]

dfs = dict(zip(names, dataframe_arr))
```

```python
products = dfs["noahs-products"]
order_items = dfs["noahs-orders_items"]
orders = dfs["noahs-orders"]
```

### Deli Items Start with DLI, Dunno which are Pasteries yet

```python
deli_order_items = order_items.merge(products[products.sku.str.contains("DLI")], on="sku")

deli_orders = orders.merge(deli_order_items, on="orderid")
deli_orders["ordered"] = pd.to_datetime(deli_orders.ordered)
deli_orders["ordered_date"] = deli_orders.ordered.dt.date
deli_orders = deli_orders.sort_values("ordered_date", ascending=True)
deli_orders
```

### Only Early Orders

```python
early_orders = deli_orders[deli_orders.ordered.dt.hour <= 5]
early_orders
```

```python
early_order_counts = early_orders.groupby("customerid").ordered.count().to_frame().reset_index().rename({"ordered": "order_ct"}, axis=1)
frequent_early_orderers = early_order_counts[early_order_counts.order_ct > 2]

orders_by_frequents = orders.merge(frequent_early_orderers, on="customerid")
orders_by_frequents
```

### The person is a bicycle repair person, so maybe they'd be buying that

```python
bike_order_items = order_items.merge(products[products.desc.str.contains("bicycle", case=False)], on="sku")
bike_orders = orders_by_frequents.merge(bike_order_items, on="orderid")
frequent_bike_orders = bike_orders.groupby("customerid").ordered.count().to_frame().reset_index().rename({"ordered": "bike_order_ct"}, axis=1).merge(frequent_early_orderers, on="customerid")

frequent_bike_orders
```

```python
customers = dfs["noahs-customers"]
customers["birthdate"] = pd.to_datetime(customers.birthdate)
young_customers = customers[customers.birthdate.dt.year > 1970]
young_customers.merge(frequent_bike_orders, on="customerid")
```

### Frequent Early Customers

```python
frequently_early = early_orders.customerid.value_counts()
frequently_early = frequently_early[frequently_early > 1].reset_index()
frequent_early_orders = early_orders.merge(frequently_early, on="customerid")
frequent_early_orders
```
